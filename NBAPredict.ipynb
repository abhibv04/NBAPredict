Using a Ridge Classifier and Sequential Feature Selector to classify and select attributes respectively. Forward Feature Selection is used

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = RidgeClassifier(alpha=1)

split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr,
                                n_features_to_select=30,
                                direction="forward",
                                cv=split,
                                n_jobs=1
                               )
```

```python
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
sfs.fit(df[selected_columns], df["target"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
#get_support() returns True if the feature selector has picked that column, and that column should be used in our model
```

Backtest to split data into seasons and use the past data to predict future results. We use data of atleast the past 2 seasons to make future predictions
Training data is the data before the season begins
Test data is the current season

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data["season"].unique())

    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = data[data["season"] < season]
        test = data[data["season"] == season]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)   #combined dataframe puts together actualvalues and predicted values
        combined.columns = ["actual", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
```

Obtaining the initial accuracy score of the model

```python
from sklearn.metrics import accuracy_score

accuracy_score(predictions["actual"], predictions["prediction"])
```

Making predictions based on average of team's performance over last 10 games- Rolling average

```python
def find_team_averages(team):
  rolling=team[selected_columns].rolling(10).mean()  #groups previous 10 rows and finds mean of the data
  return rolling
df_rolling=df_rolling.groupby(["team","season"], group_keys=False).apply(find_team_averages) #group by season and team
```

```python
rolling_cols=[f"{col}_10" for col in df_rolling.columns]  #modifying column names to _10 to indicate that they are rolling average columns
df_rolling.columns= rolling_cols
df=pd.concat([df,df_rolling],axis=1)
```

Adding in-opponent information

```python
def shift_col(team,col_name):
  next_col=team[col_name].shift(-1)   #takes next row and shifts
  return next_col
def add_col(df,col_name):
  return df.groupby("team",group_keys=False).apply(lambda x: shift_col(x,col_name))

#below is the known information to all. Need to feed the already known information, such as - [next opponent, date of next game, next home game] to model
df["home_next"]=add_col(df,"home")   #gives info whether team is at home or away next game
df["team_opp_next"]=add_col(df,"team_opp")
df["date_next"]=add_col(df,"date")
```
